```python
import random
```

```python
def random_answers(n_participants, n_questions):
    participants=[[random.randrange(-1, 2) for j in range(n_questions)] for i in range(n_participants)]
    return participants
```

```python
def create_graph(answers, n_participants, n_questions, ):
    G = nx.Graph()
    G.add_nodes_from(list(range(n_participants)))
    for i, j in itertools.combinations(range(n_participants), 2):
        G.add_edge(i, j)
    return G
```

```python
def test_weighted_match(n_participants, n_questions):
    A = random_answers(n_participants, n_questions)
    G = create_graph(A, n_participants, n_questions)
    matching = nx.max_weight_matching(G)
    nx.draw(G, with_labels = True)

    print(matching)
```

```python
test_weighted_match(20, 5)
```
