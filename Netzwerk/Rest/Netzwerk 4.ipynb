```python
import networkx as nx
import random
import itertools
import matplotlib.pyplot as plt
%matplotlib inline

n_nodes = 20
n_arr = 5
probability = 0.4


def almost_random(probability):
    chance = random.random()
    if chance <= probability:
        rand = -1
    elif chance > (1-(1-probability)/2):
        rand = 0
    else:
        rand = 1
    return(rand)

def create_arrays(n_nodes, n_arr, probability):
    arrays = [[almost_random(probability) for j in range (n_arr)] for i in range(n_nodes)]
    return arrays


def create_graph(n_nodes, n_arr, probability):
    arrays = create_arrays(n_nodes, n_arr, probability)
    G = nx.Graph()
    G.add_nodes_from(list(range(n_nodes)))
    for i, j in itertools.combinations(range(n_nodes), 2):
        for k in range(n_arr):
            if arrays[i][k] + arrays[j][k] == 1:
                G.add_edge(j,i)
    
    print(nx.draw(G, with_labels=True))
create_graph(n_nodes, n_arr, probability)
```
