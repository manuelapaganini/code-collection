Jetzt geht es daran, die Daten sichtbar zu machen.


```python
# import stuff

import pandas as pd
%matplotlib inline
```

```python
# Die Ursprungsdatei war Verzeichnis_brut. Darin habe ich mit Numbers Doppel
die Anrede umgewandelt. 

df = pd.read_csv('Verzeichnis_per.csv')
```

```python
df.head()
```

```python
df['Gender'].value_counts()
```

```python
sort_values('Approved Amount', ascending = False)
```

```python
df['Sektion'].value_counts()
```

```python
df['Abteilung'].value_counts()
```

```python
df['Funktion'].value_counts()
```

```python
df.groupby('Direktion')['Gender'].value_counts()
```

```python
df.groupby('Direktion')['Gender'].value_counts().unstack().plot(kind='bar', figsize=(12,6))
```
